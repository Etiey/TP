# Traitement d'un signal ECG

Dans cet exercice, vous allez manipuler un signal simul√© d'√©lectrocardiogramme. En v√©rit√©, le signal d'ECG n'est qu'un pr√©texte pour donner une nouvelle interpr√©tation aux fr√©quences de la transform√©e de Fourier. En effet, jusqu'√† pr√©sent, chaque transform√©e de Fourier que vous avez calcul√© √©tait associ√©e √† un vecteur de fr√©quences discr√®tes `nu` retourn√© par `sp.fft.fftfreq`. Ce vecteur de fr√©quences discr√®tes d√©coulait lui m√™me de la d√©finition du vecteur de temps discret `t` correspondant √† l'√©chelle d'observation du signal $x$. En particulier, la p√©riode d'√©chantillonnage $T_e$ de $x$ (c'est-√†-dire le pas de temps entre deux √©chantillons successifs de `t`) d√©finit la gamme de fr√©quences observables : si $x$ est compos√© de $N$ √©chantillons avec une p√©riode d'√©chantillonnage $T_e$, alors le vecteur de fr√©quences `nu` et la TFD $X$ de $x$ sont √©galement compos√©es de $N$ √©chantillons, et les fr√©quences discr√®tes s'√©talent de $-\frac{f_e}{2}$ √† $\frac{f_e}{2}$ (avec $f_e = \frac{1}{T_e}$ la fr√©quence d'√©chantillonnage de $x$).<br>

Mais comment interpr√©ter l'√©chelle des fr√©quences si `nu` n'est pas d√©finit ? (autrement dit, si les valeurs de la TFD `X[n]` ne sont rep√©r√©es que par leur indice `n`). C'est l'objet de cet exercice.

```python
# üõë D√©finition du vecteur de temps
tmax = 10 # temps max (10 secondes)
tmin = 0 #
fe = 1000 # fr√©quence d'√©chantillonnage (en Hz)
Te = 1/fe # p√©riode d'√©chantillonnage (fe points par second)
t = np.arange(tmin,tmax,Te)
```

## Simulation d'un √©lectrocardiogramme

On va regarder la transform√©e de Fourier d'un √©lectrocardiogramme (ECG) synth√©tique. La fonction `ECG_simul` prend en entr√©e le nombre de battements par minute que vous voulez simuler.

```python
# üõë Simulation d'un ECG
def ECG_simul(nb_bat,t):
    total_len = t.size # on a 10 secondes au maximum
    nb_sec = t.max()+t[1]
    Bat = np.ones(int((nb_bat*nb_sec/60)))
    fs = np.float64(total_len)/np.float64((nb_bat*nb_sec/60))
    pqrst = pywt.Wavelet('db10').dec_lo # simulation de l'ECG comme une ondelette, bref.
    ecg = np.concatenate([signal.resample(pqrst, int(r*fs)) for r in Bat])
    return ecg
```

```python
# üõë Simulation d'un ECG
Nbeats=60 # nombre de battements/minute

ecg = ECG_simul(Nbeats,t)
plt.figure(figsize=(8,5))
plt.plot(t[:ecg.size],ecg,'b',label='ECG')
plt.title('Signal ECG avec %d battements/minute'%Nbeats,fontsize=14)
plt.xlim(tmin,tmax)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.legend(loc='best',fontsize=14)
plt.show()
```
